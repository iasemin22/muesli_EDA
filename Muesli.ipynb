```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import data
orders_df = pd.read_csv("data/Orders.csv")
campaign_data_df = pd.read_csv('data/Campaign_data.csv')
intern_study_df = pd.read_csv('data/Intern_study.csv')
order_process_df = pd.read_csv('data/Order_process.csv')
orders_df.head(2)
```

```python
#make columns in lower case
orders_df.columns = orders_df.columns.str.lower()  # make column names lowercase
campaign_data_df.columns = campaign_data_df.columns.str.lower()  # make column names lowercase
order_process_df.columns = order_process_df.columns.str.lower()  # make column names lowercase
intern_study_df.columns = intern_study_df.columns.str.lower()  # make column names lowercase
```

```python
cols1 = orders_df.columns.tolist()
cols1 = [col.replace(' ', '_') for col in cols1]
orders_df.columns = cols1
cols2 = campaign_data_df.columns.tolist()
cols2 = [col.replace(' ', '_') for col in cols2]
campaign_data_df.columns = cols2
cols3 = order_process_df.columns.tolist()
cols3 = [col.replace(' ', '_') for col in cols3]
order_process_df.columns = cols3
cols4 = intern_study_df.columns.tolist()
cols4 = [col.replace(' ', '_') for col in cols4]
intern_study_df.columns = cols4
```

```python
#make columns in lower case
data3_df.columns = data3_df.columns.str.lower()  # make column names lowercase
data3_df.head(2)
```

```python
#replace spaces with _
data4_df = data3_df.copy()
cols = data3_df.columns.tolist()
cols = [col.replace(' ', '_') for col in cols]
data4_df.columns = cols
data4_df.head(2)
```
